# Pitch of missing fundamental

* pitch of harmonic complex tone is perceived to be comparable to the pitch evoked by a tone with frequency equal to the fundamental frequency of the complex tone.

* the pitch is not changed even if the lowest (fundamental) frequency is not present

* hence we can that the harmonic complex tone evokes pitch of the missing fundamental


The below given example generates a harmonic complex tone and allows you to interactively control the amplitude of the individual partials in the complex tone. You can completely remove the fundamental frequency and explore by listening whether the pitch was changed.

There are seven partials in the complex tone and the fundamental frequency is 500 Hz. Amplitude of the partials can be controled by the sliders P1 to P7. The numbers correspond to the number of partials, 1 is the fundamental frequency, 7 is the highest frequency.

!!!Be cerful not to hurt your ears, start playbeck with the volume turned down!!!


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import IPython.display as ipd

Tmax = 300e-3 # duration of the signal
F0 = 100 # fundamental frequency of the complex tone
N = 7 # number of partials
fs = 44.1e3 # sampling frequency
Nmax = round(fs*Tmax)
t = np.linspace(0,Tmax,Nmax)

def gencmplxsignal(F0,A,N,t,Tmax):  
    '''
    generate a harmonic complex tone with N partials, F0 fundamental freq and A a vector of amplitudes
    '''
    Nmax = int(np.round(Tmax*fs)) # number of samples in the signal
    
    cmplxsignal = 0
    for i in range(1,N):
        cmplxsignal += A[i]*np.sin(2*np.pi*F0*i*t)
    
    
    return 0.1*cmplxsignal
    
def update_Amplitude(P1=1,P2=2,P3=3,P4=4,P5=5,P6=6,P7=7):
    global signal    
    A = [P1, P2, P3, P4, P5, P6, P7]
    signal = gencmplxsignal(F0,A,N,t,Tmax)
    
        
    fig, ax = plt.subplots(figsize=(13, 6))
    ax.set_xlabel('time (seconds)')
    ax.set_ylabel('Amplitude')
     
    ax.plot(t,signal, lw=3, c='r')
    
    ipd.display(ipd.Audio(signal, rate=fs))
    
    


widgets.interact(update_Amplitude, P1=(0, 1, 0.05), P2=(0, 1, 0.05), P3=(0, 1, 0.05), P4=(0, 1, 0.05), P5=(0, 1, 0.05), P6=(0, 1, 0.05), P7=(0, 1, 0.05))





```
