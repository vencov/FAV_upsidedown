```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import IPython.display as ipd

Tmax = 5 # duration of the signal
fs = 44.1e3 # sampling frequency
Nmax = round(fs*Tmax) # num of samples
t = np.linspace(0,Tmax,Nmax) # time axis

L1 = -90
L2 = -80   # level of L2 tone
f1 = 1500  # f1 freq in Hz
f2 = 1600  # f2 freq in Hz

tone1 = np.sqrt(2)*10**(L1/20) * np.sin(2*np.pi*f1*t)

f2up = 1800  # final freq value of swept sine
f2down = 1500 # start freq value of swept sine
L = Tmax/np.log(f2up/f2down)
# a single tone
#tone2 = np.sqrt(2)*10**(L2/20) * np.sin(2*np.pi*f2*t)
# exponentially swept sine
tone2 = np.sqrt(2)*10**(L2/20) * np.sin(2*np.pi*f2down*L*np.exp(t/L))

signal = tone1 + tone2

ipd.display(ipd.Audio(0.001*signal, rate=fs))

plt.figure('Spectrogram',figsize=(10,10))
d, f, t, im = plt.specgram(signal, NFFT= 1024, Fs=fs, noverlap=250)
plt.ylim(0,3000)
plt.colorbar(label= "Power/Frequency")
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.show()


'''
def update_Amplitude(P1=1,P2=2,P3=3,P4=4,P5=5,P6=6,P7=7):
    global signal    
    A = [P1, P2, P3, P4, P5, P6, P7]
    signal = gencmplxsignal(F0,A,N,t,Tmax)
    
        
    fig, ax = plt.subplots(figsize=(13, 6))
    ax.set_xlabel('time (seconds)')
    ax.set_ylabel('Amplitude')
     
    ax.plot(t,signal, lw=3, c='r')
    
    ipd.display(ipd.Audio(signal, rate=fs))
    
'''


#widgets.interact(update_Amplitude, P1=(0, 1, 0.05), P2=(0, 1, 0.05), P3=(0, 1, 0.05), P4=(0, 1, 0.05), P5=(0, 1, 0.05), P6=(0, 1, 0.05), P7=(0, 1, 0.05))





```
